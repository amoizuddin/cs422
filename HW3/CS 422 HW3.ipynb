```python
def interest(suppX, suppY, suppXY):
    return "Interest", (suppXY / suppX) * suppY

def IS(suppX, suppY, suppXY):
    return "IS", suppXY / ((suppX * suppY)**0.5)

def klosgen(suppX, suppY, suppXY):
    return "Klosgen",((suppXY)**0.5) * (max(conf(suppXY, suppX) - suppY, conf(suppXY, suppY)-suppX))
    
def odds_ratio(suppXY, notXnotY, xNotY, yNotX):
    return "oddsRatio", (suppXY * notXnotY) / (xNotY*yNotX)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

#### Problem 2.1



```python
path = r"https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"
ds = pd.read_excel(path)
```

```python
ds
```

```python
France = ds["Country"] == "France"
france_ds = ds[France][["InvoiceNo", "Description"]]

france_ds[["InvoiceNo", "Description"]]
cols = france_ds["Description"].unique()
rows = france_ds["InvoiceNo"].unique()
txns = pd.DataFrame(columns=cols)
txns.insert(column="InvoiceNo", loc=0, value=rows)
txns = txns.replace(np.nan, False)
for idx, row in france_ds.iterrows():
    invoice, item= row["InvoiceNo"], row["Description"]
    txns.loc[txns['InvoiceNo'] == invoice, item] = True

txns = txns.drop("InvoiceNo", axis=1)
```

```python
freq_itemsets = apriori(txns, min_support=0.05, use_colnames=True)
freq_itemsets
```

```python
rules_confidence = association_rules(freq_itemsets, metric='confidence',min_threshold=1)
rules_confidence
```

```python
rules_lift = association_rules(freq_itemsets, metric='lift',min_threshold=11)
rules_lift
```

The rule with highest confidence is not the same as the rule with the highest lift. Confidence is a measure of how often {B} occurs with {A}. In the Fomula is supportcount(A U B) / supportcount(A). Lift adds a factor of 1/supportcount(B). Thus if B occured with A every time the lift would be equal 1. But as B increases lift increases also which means that {A} -> {B} is an actual rule and {B} isnt just a frequent item.

```python
name1 = 'Chocolate Coffee'
name2 = 'Chocolate Cake'

cnc = bakery_ds[[name1,name2]]

condition1 = cnc[name1]==1
condition2 = cnc[name2]==1

data = cnc.groupby([condition1,condition2]).count()
data
```

```python
print(cnc[name1].corr(cnc[name2]))
print(cnc[name2].corr(cnc[name1]))
```

These items are symmetric binary variables. The association rules have the same corrolation values because to calculate it we do txns where both items occur multiplied by txns where neither items occur then subtract the product of the occurances for each item. then you divite that by the root of the sum of total occurances and not-occurances for each item. As seen order doesnt matter for corrolation as corrolation compares the items to each other.
